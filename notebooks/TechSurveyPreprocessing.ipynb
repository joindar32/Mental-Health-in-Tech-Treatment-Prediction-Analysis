##Удаление дубликатов и обработка пропусков.
Удаляем дубликаты строк, где все значения совпдаают. Далее смотрим на пропущенные значения. Мы заполняем пропуски в столбце state, потому что это применимо далеко не ко всем странам, в таком случае мы просто заменяем NaN на "Not Applicable". У нас довольно мало строк с пропуском в столбце self_employed, поэтому можем удалить такие строчки не опасаясь за искажение данных. В столбцах work_interfere и comments мы просто заменяем пустые значения на "No data" и "No comments" соответсвенно. После заполнения пропусков проверяем, чтобы их больше не осталось.


```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['state'] = df["state"].fillna("Not Applicable")
df = df.dropna(subset=['self_employed'])
df['work_interfere'] = df['work_interfere'].fillna("No data")
df['comments'] = df['comments'].fillna("No comments")
```

```python
df.isnull().sum()
```

##Стандартизация значения в столбец "Gender"
Для того, чтобы легче было работать с полем "Gender",который содержит очень много разных значений, группируем все ответы в 3 возможные категории: male, female, other. Новый столбец будет называться "Standardized_gender"

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].str.lower()
gender_mapping = {
    'male' : ['male', 'm', 'male-ish', 'maile', 'make', 'mal', 'man', 'msle',
             'mail', 'malr', 'something kinda male?',
             'cis male', 'male (cis)', 'guy (-ish) ^_^', 'male ', 'cis man',
             'ostensibly male, unsure what that really means'],
    'female' : ['female', 'female ', 'f', 'woman', 'femake', 'female (cis)',
               'femail', 'cis female', 'cis-female/femme'],
    'other' : ['trans-female', 'something else', 'non-binary', 'neuter',
              'female (trans)', 'queer/she/they', 'nah', 'all', 'enby', 'fluid',
              'genderqueer', 'androgyne',
              'agender', 'male leaning androgynous', 'trans woman', 'queer',
              'a little about you', 'p']
}

df['Standardized_gender'] = df['Gender'].apply(lambda x: next((key for key, value in gender_mapping.items() if x in value), x))

```

```python
df['Standardized_gender'].value_counts()
```

##Обработка аномальных данных в поле "Age"
Обрабатываем столбец возраста, оставляя только значения в диапазоне от 18 до 100 лет, чтобы удалить аномальные значения.

```python
df['Age'].unique()
```

```python
df = df[(df['Age'] > 18) & (df['Age'] < 100)]
```

##Сохранение очищенного датасета
Сохраняем очищенный датасет в "cleaned_df.csv" для дальнейшей работы с ним на этапе анализа данных

```python
df.to_csv('cleaned_df.csv', index=False)
files.download('cleaned_df.csv')
```
